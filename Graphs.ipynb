```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/NINO_NEW.xlsx')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['NINO3.4'], color='b', linewidth=2, marker='o', markersize=4)
plt.title('NINO3.4 Values Variation (1950-2023)')
plt.xlabel('Date')
plt.ylabel('NINO3.4 Value')
plt.grid(True)
plt.show()
```

```python
df['Year'] = df['Date'].dt.year
df
```

```python
enso_by_year = df.groupby('Year')['NINO3.4'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(enso_by_year['Year'], enso_by_year['NINO3.4'], marker='o', linestyle='-', color='b')
plt.title('Average NINO3.4 Values per Year (1950-2023)')
plt.xlabel('Year')
plt.ylabel('Average NINO3.4 Value')
plt.grid(True)
plt.show()
```

```python
df2 = pd.read_excel('/content/merged.xlsx', index_col=0)
df2
```

```python
correlation = df2['NINO3.4'].corr(df2['swh'])
print(f"Correlation between ENSO and Significant Wave Height: {correlation:.2f}")
```
